```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import tweepy
import json
from config import (consumer_key, 
                    consumer_secret, 
                    access_token, 
                    access_token_secret)
from datetime import datetime
```

```python
# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# Target Users
target_users = ("@BBCWorld", "@CBSNews", "@CNN", "@FoxNews", "@NYTimes")

# Variables for holding sentiments
sentiments = []

# Loop through target users
for user in target_users:
    
    # Counter
    counter = 0
    
    # Get tweets from each user's home feed
    public_tweets = api.user_timeline(user, count = 100)
    
    # Loop through tweets
    for tweet in public_tweets:
        
        # Vader analysis
        compound = analyzer.polarity_scores(tweet["text"])["compound"]
        positive = analyzer.polarity_scores(tweet["text"])["pos"]
        negative = analyzer.polarity_scores(tweet["text"])["neg"]
        neutral = analyzer.polarity_scores(tweet["text"])["neu"]
        
        # Other info for dictionary
        tweet_number = counter
        date = tweet["created_at"]
        tweet_text = tweet["text"]
        
        
        # Add sentiments to each tweet to empty list
        sentiments.append({"Twitter": user,
                          "Date": date,
                          "Tweet": tweet_text,
                          "Tweets Number": counter,
                          "Compound": compound,
                          "Positive": positive,
                          "Neutral": neutral,
                          "Negative": negative})
        
        # Add to counter
        counter = counter + 1
        
        
    
```

```python
# Check to make sure I have 500 sentiments (100 for each news org)
len(sentiments)
```

```python
# Create dataframe
news_tweets_df = pd.DataFrame(sentiments)
news_tweets_df.head()
```

```python
# Organize data frame
news_df = news_tweets_df[["Twitter", "Date", "Tweets Number", "Tweet", "Compound", "Positive", "Neutral", "Negative"]]
news_df = news_df.rename(columns={"Twitter": "News Organization", "Tweets Number": "Tweets Ago"})
news_df.head(10)
```

```python
# Output data frame to csv
news_df.to_csv("News_Tweets.csv")
```

### Sentiment Plot of Last 100 Tweets by News Organization

```python
# Show compound score by organization for last 100 tweets
org_compound = news_df.pivot(index="Tweets Ago", columns="News Organization", values="Compound")
org_compound.head(10)
```

```python
# Create scatter plot
x = np.arange(100)

plt.scatter(x, org_compound["@BBCWorld"])
plt.scatter(x, org_compound["@CBSNews"])
plt.scatter(x, org_compound["@CNN"])
plt.scatter(x, org_compound["@FoxNews"])
plt.scatter(x, org_compound["@NYTimes"])

# Plot labels
plt.title("Sentiment Analysis of Media Tweets (09-17-2018)")
plt.xlabel("Tweets Ago")
plt.ylabel("Tweet Polarity")
plt.legend(loc="best", bbox_to_anchor=(1,1))
plt.grid(True)
plt.ylim(-1,1)
plt.xlim(103, -3)

# Save Plot
plt.savefig("Sentiment_Analysis.png")
plt.show()
            
```

### Overall Sentiment of News Organization

```python
# Get the average compound sentiment of each news org
group_df = news_df.groupby("News Organization")
group_mean = group_df["Compound"].mean()
group_mean
```

```python
# Create graph
plt.bar("@BBCWorld", group_mean["@BBCWorld"])
plt.bar("@CBSNews", group_mean["@CBSNews"])
plt.bar("@CNN", group_mean["@CNN"])
plt.bar("@FoxNews", group_mean["@FoxNews"])
plt.bar("@NYTimes", group_mean["@NYTimes"])

# Graph labels
plt.title("Overall News Organization Sentiment (09-17-2018)")
plt.xlabel("News Organization")
plt.ylabel("Tweet Polarity")
plt.ylim(-0.12, 0.12)

# save graph
plt.savefig("Overall_Sentiment.png")
plt.show()
```

## Analysis

* Overall all news organizations had a negative average polarity for the last 100 tweets.
* Out of the five news organizations, BBCWorld was the most negative and CBSNews was the least.
* The negative sentiments make me wonder if news organizations are using fear-based tactics to draw people's attentions or are there just that many bad things going on in the world?
